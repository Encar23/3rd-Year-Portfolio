```python
%matplotlib inline
import numpy as np
from pylab import *
from scipy import stats

pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3
scatter(pageSpeeds, purchaseAmount)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount)
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
```

## Try increasing the random variation in the test data and see what effect it has on the r-squared error value

```python
pageSpeeds = np.random.normal(3.0, 5.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(4.0, 3, 1000)) * 3
scatter(pageSpeeds, purchaseAmount)

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)

fitLine = predict(pageSpeeds)

plt.scatter(pageSpeeds, purchaseAmount, color='k')
plt.plot(pageSpeeds, fitLine, color='y')
plt.show()
```

# Conclusion

We have observed that, as standard deviation increases, random variation increases as well, therefore a higher standard deviation leads to bigger differences and lower accuracy of the fitted values. 
Since higher r-squared error values represent smaller differences and higher accuracy between the observed test data and the fitted values, we can conclude that,   standard deviation is inversely proportional to the r-squared error value.
